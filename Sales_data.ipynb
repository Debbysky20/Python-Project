```python
grouped = sales_data.groupby('Gadget')['Quantity'].sum().sort_values()
grouped.head()
```

```python
plt.figure(figsize=(10,6))
plt.bar(grouped.index, grouped.values, color=['blue', 'orange'])
plt.xlabel('Gadget')
plt.ylabel('Quantity')
plt.title('Total_Quantity by Gadget')
plt.show()
```
